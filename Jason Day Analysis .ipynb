```python
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv('data\pgadata_2010_2018.csv', index_col='Player Name')
pd.set_option('display.max_columns', 17)
pd.set_option('display.max_rows', 20)
```

```python
df['Wins'].fillna(0, inplace=True)
df['Top 10'].fillna(0, inplace=True)
```

```python
jason_day = df.loc['Jason Day'].sort_values(by='Year')
```

```python
jason_day
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
jd_avg_score = jason_day['Average Score']

    
plt.plot(years, jd_avg_score)
plt.ylabel('Average Score')
```
